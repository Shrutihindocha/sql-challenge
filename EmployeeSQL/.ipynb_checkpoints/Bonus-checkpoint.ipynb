```python
# Dependencies
# ----------------------------------
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Create engines
engine = create_engine('postgresql://postgres:postgres@localhost:5432/sql-challenge')
```

```python
#Creating dfs for salaries, titles and employees

salaries_df = pd.read_sql("SELECT * FROM salaries", engine)
#salaries_df
titles_df = pd.read_sql("SELECT * FROM titles", engine)
# titles_df
employees_df = pd.read_sql("SELECT * FROM employees", engine)
#employees_df
```

```python
#Renaming title_id to emp_title_id

titles_df = titles_df.rename(columns={'title_id': 'emp_title_id'})

titles_df
```

```python
#merging salaries and title IDs into the employees table 
merged_df = employees_df.merge(salaries_df, on="emp_no", how="inner")

merged_df = merged_df.merge(titles_df, on="emp_title_id", how = "inner")

merged_df
```

```python
#group by title

title_salary_df = merged_df.groupby("title").mean()["salary"]

title_salary_df = title_salary_df.reset_index()

title_salary_df
```

```python
#plotting bar chart for salaries by title
x_axis = title_salary_df["title"]
y_axis = title_salary_df["salary"]

plt.xticks(rotation=90)
plt.bar(x_axis, y_axis, color="b", align="center")
```
